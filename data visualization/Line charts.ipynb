```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("C:/Users/HP/Desktop/Kaggle Assignment/Data Visualization/Line Charts/museum_visitors.csv",index_col=False)
df=pd.DataFrame(data)
df.head()
```

```python
#How many visitors did the Chinese American Museum receive in July 2018?
#greater than the start date and smaller than the end date
ca_museum_jul18 = df[(df['Date'] == '2018-07-01')]
ca_museum_jul18 = ca_museum_jul18['Chinese American Museum'].values
ca_museum_jul18
```

```python
#In October 2018, how many more visitors did Avila Adobe receive than the Firehouse Museum?

avila_oct18=df[(df['Date'] =='2018-10-01')]
avila_oct18=(avila_oct18['Avila Adobe'] - avila_oct18['Firehouse Museum']).values
avila_oct18
```
